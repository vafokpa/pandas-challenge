```python
# Dependencies and Setup
import pandas as pd
import os
import numpy as np

# File to Load (I used a drop_duplicate at some point and messed up the csv file so I found another file )
path = os.path.join("Resources", "purchase_data1.csv")

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(path)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Creating a variable to hold the total numbers of players
total_players = purchase_data["SN"].nunique()
total_players

total_players_df= pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Creating a Variable for unique items
unique_items = purchase_data["Item ID"].nunique()

# Creating a Variable for average price
average_price = purchase_data["Price"].mean()

# Creating a Variable for purchases
num_purchases = purchase_data["Purchase ID"].count()

# Creating a Variable for Total Revenue
revenue = purchase_data["Price"].sum()

# Creating a Dataframe
summary_df = pd.DataFrame({"Number of Unique Items": [unique_items], "Average Price": ["$" +str(round(average_price,2))],"Number of Purchases":[num_purchases], "Total Revenue": ["$" + str(revenue)]})
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Create a dataframe for just the male demographic
male_df = purchase_data[purchase_data["Gender"] == "Male"]
male = male_df["SN"].nunique()

# Create a dataframe for just the female demographic
female_df = purchase_data[purchase_data["Gender"] == "Female"]
female = female_df["SN"].nunique()

# Create a dataframe for just the Other / Non-Disclosed demographic
other_df = purchase_data[purchase_data["Gender"] == "Other / Non-Disclosed"]
other = other_df["SN"].nunique()

# Create a dataframe
gender_demodf = pd.DataFrame({"Gender": ["Male", "Female", "Other / Non-Disclosed"], "Total Count": [male, female, other], "Percentage":[str(round(male/total_players * 100,2))+"%", str(round(female/total_players * 100,2)) +"%", str(round(other/total_players * 100,2))+ "%"]})
gender_demodf
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create a goup by Gender
group_df = purchase_data.groupby(["Gender"])
print(group_df)

# Count groupby
gender_count = group_df["Purchase ID"].count()

# Average price groupby
ave_price = group_df["Price"].mean()

# Total purchase value groupby
tot_purchase = group_df["Price"].sum()

# Average total purchase per person groupby
ave_totalpur = group_df["Price"].sum() / group_df["SN"].nunique()

# Putting it all in a DataFrame
gender_analysisdf = pd.DataFrame({"Purchase Count": gender_count, "Average Purchase Price": round(ave_price,2).map("${:.2f}".format), "Total Purchase Value": tot_purchase.map("${:.2f}".format), "Average Total Purchase per Person": round(ave_totalpur,2).map("${:.2f}".format)})
gender_analysisdf
```

# Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Creating a copy of the data frame without duplicate buyers
purchase_copy = purchase_data.copy()
purchase_copy = purchase_copy.drop_duplicates(subset = "SN", keep = "first")
```

```python
# Create the bins in which Data will be held
bins = [0,9, 14, 19, 24, 29, 34, 39, 100]

# Create the names for the four bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39","40+"]

# Add bins to dataframe
purchase_copy["Age Group"] = pd.cut(purchase_copy["Age"], bins, labels=group_names)

# Create a groupby Age
age_df = purchase_copy.groupby(["Age Group"])
print(age_df)

# Create the count values
age_count = age_df["Purchase ID"].count()

# Create the percentage values
age_percent = round(age_df["Purchase ID"].count()/total_players * 100,2)

# Create the dataframe
summary_age_df = pd.DataFrame({"Total Count": age_count, "Percentage of Players (%)": age_percent.map("{:.2f}%".format)})
summary_age_df

```

## Purchasing Analysis (Age)

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Groupby spenders
top_spendersdf = purchase_data.groupby(["SN"])
print(top_spendersdf)
top_spendersdf

# Creating Purchase Count Values
spender_count = top_spendersdf["Purchase ID"].count()

# Creating Average purchase price
spender_avg = top_spendersdf["Price"].mean()

# Creating Total Purchase Value
spender_tot = top_spendersdf["Price"].sum()

#Creating dataframe
summary_topspender_df = pd.DataFrame({"Purchase Count": spender_count, "Average Purchase Price": round(spender_avg,2).map("${:.2f}".format), "Total Purchase Value": spender_tot.map("${:.2f}".format)})
summary_topspender_df.sort_values(by = 'Total Purchase Value', ascending = False).head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# GroupBy Item ID and Item name
popular_items = purchase_data.groupby(["Item ID","Item Name"])
print(popular_items)

# Create values for purchase count
popular_count = popular_items["Item Name"].count()

# Create values for item price
popular_price = popular_items["Price"].unique()

# Because this was created as a list, I use a list comprehension to change the format (instead of .map in the dataframe)
popular_price = ["$"+str(float(i)) for i in popular_price]

# Create values for total purchase value
popular_total = popular_items["Price"].sum()

# Create Dataframe for summary information
popular_items_df = pd.DataFrame({"Purchase Count": popular_count, "Item Price": popular_price, "Total Purchase Value": popular_total.map("${:.2f}".format)})
popular_items_df.sort_values(by = 'Purchase Count', ascending = False).head()
```
